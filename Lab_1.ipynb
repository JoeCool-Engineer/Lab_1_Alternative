### 1.1 Vector Space Properties
A vector space is a collection of vectors that we can add and multiply by scalars. Let’s explore these operations through examples and visualization.

### Exercise 1.1: Vector Operations in ℝ^2

```python
## Setup
import numpy as np
import matplotlib.pyplot as plt

def setup_plot():
    plt.figure(figsize=(10, 6))
    plt.grid(True)
    plt.axhline(y=0, color='k', linestyle=':')
    plt.axvline(x=0, color='k', linestyle=':')


## Vector Operations
def plot_vector(v, color='blue', label=None):
    """Plot a 2D vector from origin"""
    plt.quiver(0, 0, v[0], v[1],
              angles='xy', scale_units='xy', scale=1,
              color=color, label=label)

def vector_addition(v1, v2):
    """Demonstrate vector addition"""
    setup_plot()
    plot_vector(v1, 'blue', 'v1')
    plot_vector(v2, 'red', 'v2')
    plot_vector(v1 + v2, 'green', 'v1 + v2')
    plt.title('Vector Addition')
    plt.legend()
    plt.show()

def vector_subtraction(v2, v3):
    """Demonstrate vector subtraction"""
    setup_plot()
    plot_vector(v2, 'blue', 'v2')
    plot_vector(v3, 'red', 'v3')
    plot_vector(v2 - v3, 'green', 'v2 - v3')
    plt.title('Vector Subtraction')
    plt.legend()
    plt.show()

## Examples
# Example vectors
v1 = np.array([1, 2])
v2 = np.array([3, 1])
v3 = np.array([-1, 4])

# Basic operations
print(f"v1 + v2 = {v1 + v2}")
print(f"2v1 = {2 * v1}")
print(f"v2 - v3 = {v2 - v3}")

# Visualize
vector_addition(v1, v2)
vector_subtraction(v2, v3)

## Practice Problems
def practice():
    """Vector operation practice problems"""
    # Your code here
    pass
```

```python
def plot_vector(v, color='blue', label=None):
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color=color, label=label)
    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.grid(True)
    if label:
        plt.legend()
        
# Example usage
plot_vector(v1, 'red', 'v1')
plot_vector(v2, 'blue', 'v2')
plt.show()
```

```python
ex_1 = np.array([2, 1])
ex_2 = np.array([1, -2])
ex_b = np.array([3, -1])

plot_vector(ex_1, 'red', 'ex_1')
plot_vector(ex_2, 'blue', 'ex_2')
plot_vector(ex_b, 'green', 'ex_b')
plt.show()

ex_b_transpose = ex_b.transpose()
ex_3 = np.array([[2, 1], [1, -2]])
ex_c = np.array([[3], [-1]])
ex_3_transpose = ex_3.transpose()
linear_combination = np.dot(ex_3.transpose(), ex_c)
print(linear_combination)
plot_vector(linear_combination, 'orange', 'linear_combination')
plt.show()



```

```python
# Define the equation y = mx + b
m1 = -2  # Slope 1
b1 = 3  # Y-intercept 1

m2 = .5  # Slope 2
b2 = .5  # Y-intercept 2

# Create x values using numpy's linspace for a smooth line
x1 = np.linspace(-5, 5, 400)
x2 = np.linspace(-5, 5, 400)

# Calculate corresponding y values
y1 = m1 * x1 + b1
y2 = m2 * x2 + b2

# Create the plot
plt.plot(x1, y1, label=f'2x + y = 3')
plt.plot(x2, y2, label=f'x - 2y = -1')
plt.plot(1, 1, 'o', label='Intersection (1,1)')

# Add labels and title
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Graph of a Linear Equation')

# Add a grid for better readability
plt.grid(True)

# Add a legend
plt.legend()

# Display the plot
plt.show()
```
